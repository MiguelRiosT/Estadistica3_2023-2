# Gr√°fico de 2D y 1D

```python
# Entrenar PCA para 2D
pca = PCA(n_components=2)
pca_result_2d = pca.fit_transform(DATOS_sc)

# Gr√°fico de dispersi√≥n 2D
plt.figure(figsize=(8, 6))
plt.scatter(pca_result_2d[:, 0], pca_result_2d[:, 1], alpha=0.7)
plt.title('Gr√°fico de dispersi√≥n PCA en 2D')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()

# Entrenar PCA para 1D
pca = PCA(n_components=1)
pca_result_1d = pca.fit_transform(DATOS_sc)

# Gr√°fico de dispersi√≥n 1D
plt.figure(figsize=(8, 4))
plt.scatter(pca_result_1d, np.zeros_like(pca_result_1d), alpha=0.7)
plt.title('Gr√°fico de dispersi√≥n PCA en 1D')
plt.xlabel('Componente Principal 1')
plt.yticks([])
plt.show()
```

```python
# Calcular la matriz de covarianza
cov_matrix = np.cov(DATOS_sc.T)

# Calcular los eigenvectores y eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Gr√°fico de dispersi√≥n 2D con eigenvectores
plt.figure(figsize=(8, 6))
plt.scatter(pca_result_2d[:, 0], pca_result_2d[:, 1], alpha=0.7)
plt.title('Gr√°fico de dispersi√≥n PCA en 2D con Eigenvectores')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')

# Dibujar los eigenvectores como flechas
for i, (value, vector) in enumerate(zip(eigenvalues, eigenvectors.T)):
    arrow_props = dict(facecolor='black', arrowstyle='->')
    plt.annotate(f'Eigenvalue {i+1}', xy=(0, 0), xytext=(vector[0], vector[1]), arrowprops=arrow_props)

plt.show()
```
